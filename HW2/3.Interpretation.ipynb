Все методы близко сошлись в количестве кластеров: 8.
Можно предположить, что 8 это число категорий, которые были заложены создателями набора данных в поле Purpose, что выглядит убедительно и логично. Уточним предположение с помощью расчета средних значений признаков для каждого из кластеров и изучим Box Plots по кластерам с помощью программы Jamovi.

Результаты сохранены в PDF файлы. На диаграммах черная прямоугольная метка означает среднее значение.

## Hierarhical clustering

german_credit_data_Labeled_Hierarhical.pdf

Иерархический алгоритм тоже опирался на признак Purpose, но не прямо, а в сочетаниях + на пол:
- Категория, где брали кредит на автомобиль разделена на два подкласса: 0, 5.
- Категория, где брали кредит на мебель разделена на два подкласса: 4, 7.
- Категория, где брали кредит на ТВ/радио разделена на два подкласса: 6, 8.

Что можно сказать про подклассы:
0. Цель: автомобиль. В основном, мужчины.
4. Цель: на мебель. В основном, женщины.
5. Цель: автомобиль. В основном, женщины.
6. Цель: на ТВ/радио. В основном, женщины.
7. Цель: на мебель. В основном, мужчины.
8. Цель: на ТВ/радио. В основном, мужчины.

Другие кластеры:
1. Цель: на бизнес. В основном это люди, владеющие жильём. 
2. Цель: домашнее обустройство, ремонт, отпуск/другое — предположим, что здесь больше не про отпуск, а про "другое", но отпуск тоже логично вписывается, ведь ремонт квартиры удобно делать в отпуск. Вместе такое сочетание категорий можно назвать как: увлеченный ремонт квартиры/дома. В основном это люди, владеющие жильём.
3. Цель: на учёбу. 


## K-means

german_credit_data_Labeled_KMeans.pdf

K-means опирался на признак Purpose в сочетаниях + на пол:
- Категория, где брали кредит на автомобиль разделена на два подкласса: 2, 6.
- Категория, где брали кредит на мебель разделена на два подкласса: 3, 7.
- Категория, где брали кредит на ТВ/радио разделена на два подкласса: 1, 5.

Что можно сказать про подклассы:
1. Цель: на ТВ/радио. Мужчины.
2. Цель: на автомобиль. Мужчины.
3. Цель: на мебель. Женщины.
5. Цель: на ТВ/радио. Женщины.
6. Цель: на автомобиль. Женщины.
7. Цель: на мебель. Мужчины.

Другие кластеры:
0. Цель: обустройство дома, обучение, ремонт, отпуск/другое. Можно предположить: работающие или обеспеченные родителями студенты старших курсов или начинающие специалисты.
4. Цель: на бизнес. В основном: женщины — бизнес-леди.
