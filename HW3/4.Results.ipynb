## Действительно ли модели справляются со своей задачей 
И на сколько хорошо они это делают.

### DBScan

DBScan не смог отработать на полном наборе данных: не хватило 64 Гб ОЗУ — производительность была низкая и закончились опыты переполнением ОЗУ.

Поэтому выборка была значительно уменьшена (со стратификацией по Class).

Это не позволяет нам сравнивать алгоритмы между собой, но, грубо оценим результат по полученным данным.

Удалось точно подобрать гиперпараметр eps. Процент аномалий ожидается в данных — 0.17 %

Для достижения детектирования доли аномалий: 0.0017

Подобран гиперпараметр eps: 20.5

#### Тренировочные данные
По диаграмме pairplot DBScanPairplot.png мы видим, что модель определила выбросы, но многие удаленные точки похожие на связанные тренды модель не отнесла к аномалиям. Можно заметить, что модель почти не отнесла шумовые точки на границах скоплений к аномальным.

#### Тестовые данные
По диаграмме UMAP DBScanUMAPpredictedAnomaliesAtTest.png мы видим, что определенные аномалии не выделяются от общей массы данных.

По classification report и confusion matrix мы видим, что модель не смогла определить реальные случаи вторжений.
Очевидно, что на отрицательный результат модели повлияло сильное ограничение объема данных.

### Isolation Forest

Модель точно обработала гиперпараметр: какой процент аномалий ожидается в данных — 0.17 %

Количество выбросов в выборке: 388

Процент выбросов в выборке 0.17%

#### Тренировочные данные
По диаграмме pairplot IsolationForestPairplot.png мы видим, что модель хорошо определила выбросы, но захватила в выбросы и часть пограничных точек.

#### Тестовые данные
По диаграмме UMAP IsolationForestUMAPpredictedAnomaliesAtTest.png мы видим, что модель часть аномалий выделила убедительно, а часть смешалась с основной массой данных.

По classification report и confusion matrix мы видим, что класс 0 определился отлично, но класс 1 определился с качеством f1 = 0.25, т.е. плохо.
Получается, что модель плохо срабатывает на факты вторжений.

Мы видим, что определенные аномалии нельзя уверенно отделить от общей массы данных.

### Isolation Forest

Мы видим, что значительную часть аномальных точек нельзя отделить от общей массы данных или они находятся на границе областей pairplots. Получается, что аномалии часто замаскированы и их трудно уверенно отделить от обычных транзанкций. Модель смогла правильно среагировать на аномалии только для 26% аномальных случаев.
